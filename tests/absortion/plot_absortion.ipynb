```python
from luminis_mc import ResultsLoader
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def plot_absortion_image(absorption_data, abs_r_len, abs_z_len, title="Absoriton"):
    plt.rcParams['text.usetex'] = True
    plt.figure(figsize=(6, 5))
    extent = [0.0, abs_z_len, -abs_r_len, abs_r_len]
    plt.imshow(absorption_data, cmap='inferno', origin='lower', extent=extent, aspect='auto')
    plt.colorbar(label=r'Absorbed Fraction [$\mathrm{cm}^{-3}$]')
    plt.title(title)
    plt.xlabel('Depth [$1/l$]')
    plt.ylabel('Radius [$1/l$]')
    plt.grid(False)
    plt.show()
```
