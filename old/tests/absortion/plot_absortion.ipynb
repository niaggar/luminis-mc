```python
from luminis_mc import ResultsLoader
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
def plot_absortion_image(absorption_datas, indexes, abs_r_len, abs_z_len, title="Absoriton"):
    fig = plt.figure(figsize=(16, 10))
    fig.suptitle(f"{title}", fontsize=16)

    n_cols = 2
    n_rows = (len(indexes) + n_cols - 1) // n_cols

    for i, idx in enumerate(indexes):
        absorption_data = absorption_datas.absorption_image("absorption", idx)

        ax = fig.add_subplot(n_rows, n_cols, i + 1)
        extent = [0.0, abs_z_len, -abs_r_len, abs_r_len]
        ax.imshow(absorption_data, cmap='inferno', origin='lower', extent=extent, aspect='auto')
        ax.set_title(f"t={idx * 1.0:.1f} ns")
        ax.set_xlabel("Depth $[l]$")
        ax.set_ylabel("Radius $[l]$")

    plt.tight_layout()
    plt.show()
```

```python
base_dir = "/Users/niaggar/Documents/Thesis/Progress/02Mar26/"
dir_name = "absorption_1-layer"
data = ResultsLoader(base_dir + dir_name)

# Anysotropy g1=0.9241036966036794
meta_sensor_fluence = data.sensor_meta("planarfluence")
meta_absorption = data.absorption_meta("absorption")
indexes = [3, 9, 15, 17, 29, 39]

plot_absortion_image(data, indexes, meta_absorption['radius'], meta_absorption['depth'], title="Absorption Image for 1-layer medium")
```

```python
base_dir = "/Users/niaggar/Documents/Thesis/Progress/02Mar26/"
dir_name = "absorption_2-layer"
data = ResultsLoader(base_dir + dir_name)

# Anysotropy g1=0.9241036966036794
# Anysotropy g2=0.0044753415355474
meta_sensor_fluence = data.sensor_meta("planarfluence")
meta_absorption = data.absorption_meta("absorption")
indexes = [3, 9, 15, 17, 29, 39]

plot_absortion_image(data, indexes, meta_absorption['radius'], meta_absorption['depth'], title="Absorption Image for 2-layer medium")
```
